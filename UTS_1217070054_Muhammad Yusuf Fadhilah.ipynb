10. Setelah melakukan konversi BGR ke HSV kemudian tampilkan histogram untuk Band Merah/Red Channel, Hijau/Green Channel dan Biru/Blue channel (boleh menggunakan Matplotlib atau cv2.imshow) (Poin: 10)




```python
#Muhammad Yusuf Fadhilah
#1217070054

import cv2
import matplotlib.pyplot as plt

img = cv2.imread("C:/Users/muham/OneDrive/Dokumen/PEMROGRAMAN/Visual Studio/Python/opencv/2/gura.jpeg")

hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

histr = cv2.calcHist([hsv], [0], None, [256], [0, 256])
histg = cv2.calcHist([hsv], [1], None, [256], [0, 256])
histb = cv2.calcHist([hsv], [2], None, [256], [0, 256])

plt.plot(histr, color = 'red', label='R')
plt.plot(histg, color = 'green', label='G')
plt.plot(histr, color = 'blue', label='B')

plt.title('Histogram RGB')
plt.legend()
plt.show()
```

12. Menaikan brighness dengan menambahkan nilai konstanta = 100. Tampilkan gambar sebelum brightness dan sesudah (poin: 10)

```python
#Muhammad Yusuf Fadhilah
#1217070054

import cv2
import numpy as np

img = cv2.imread("C:/Users/muham/OneDrive/Dokumen/PEMROGRAMAN/Visual Studio/Python/opencv/2/gura.jpeg")

hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

value = 10000000

hsv[:,:,2] = cv2.add(hsv[:,:,2], value)
img = cv2.cvtColor(hsv, cv2.COLOR_HSV2BGR)
cv2.imshow('Gura Terang', img)
cv2.waitKey(0)
cv2.destroyAllWindows
```

13. Putar gambar 90 derajat dan tampilkan (poin:10)

```python
#Muhammad Yusuf Fadhilah
#1217070054

import cv2
import numpy as np

img = cv2.imread("C:/Users/muham/OneDrive/Dokumen/PEMROGRAMAN/Visual Studio/Python/opencv/2/gura.jpeg")

(height, width) = img.shape[0:2]
mutermetrik = cv2.getRotationMatrix2D((width/2, height/2), -90 ,0.5)
mutergambar = cv2.warpAffine(img, mutermetrik, (width, height))
cv2.imshow('Gura Salto', mutergambar)
cv2.waitKey(0)
cv2.destroyAllWindows
```

13. Normalisasikan gambar dan tampilkan historgramnya (poin 5)
